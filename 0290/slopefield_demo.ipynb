```python
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (14,8)
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-10,10,50)
y = np.linspace(-5,5,25)
```

```python
def diffy1(t,y):
    return y
def diffy2(t,y):
    return -2*y*t
```

```python
for j in x:
    for k in y:
        slope = diffy1(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')
        
plt.plot(np.linspace(-10,10,80),np.exp(np.linspace(-10,10,80)))            # Increasing Solutions
plt.plot(np.linspace(-10,10,80),2*np.exp(np.linspace(-10,10,80)))
plt.plot(np.linspace(-10,10,80),.2*np.exp(np.linspace(-10,10,80)))

plt.plot(np.linspace(-10,10,80),-np.exp(np.linspace(-10,10,80)))           # Decreasing Solutions
plt.plot(np.linspace(-10,10,80),-2*np.exp(np.linspace(-10,10,80)))
plt.plot(np.linspace(-10,10,80),-.2*np.exp(np.linspace(-10,10,80)))

plt.plot(np.linspace(-10,10,80),0*np.linspace(-10,10,80))          # Constant Solution


plt.ylim(ymax=5,ymin=-5)
plt.xlim(xmax=10,xmin=-10)        
plt.title("Slope Field y'")
plt.grid(True)
```
