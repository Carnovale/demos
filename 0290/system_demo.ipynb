```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

**Euler-System**

```python
N = 50
t0 = 0
tf = 3*np.pi
y0 = -1
v0 = -2
def yp(t,y,v):
    return v
def vp(t,y,v):
    return -y
```

```python
def euler_sys(t0,tf,y0,v0,yp,vp,N):
    ts = [t0]
    ys = [y0]
    vs = [v0]
    delta = (tf - t0)/float(N)
    for i in range(N):
        y0temp = y0
        y0 += delta*yp(t0,y0,v0)
        v0 += delta*vp(t0,y0temp,v0)
        t0 += delta
        ts = np.append(ts,t0)
        ys = np.append(ys,y0)
        vs = np.append(vs,v0)
    return ts, ys, vs
```

```python
ts, ys, vs = euler_sys(t0,tf,y0,v0,yp,vp,1000)
fig, axarr = plt.subplots(2)
axarr[0].plot(ys,vs)
axarr[1].plot(ts,ys,ts,vs)
```

```python
def p(t):
    return 1
def q(t):
    return t
def f(t):
    return np.cos(t)
```

```python
def yp(t,y,v):
    return v
def vp(t,y,v):
    return f(t) - p(t)*v - q(t)*y
```

```python
ts, ys, vs = euler_sys(t0,tf,y0,v0,yp,vp,10000)
fig, axarr = plt.subplots(2)
axarr[0].plot(ys,vs)
axarr[1].plot(ts,ys,ts,vs)
```
